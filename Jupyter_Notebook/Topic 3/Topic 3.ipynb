```python
data = np.genfromtxt("live-birth.csv",
            delimiter=',',
            names=True, dtype=(int, "U7", "U7", int))

birth_count = []
data_titles = np.unique(data['year'])

for i in range(len(data_titles)):
    data_year = data[data['year'] == data_titles[i]]['birth_count'].sum()
    birth_count.append(data_year)

plt.bar(data_titles, birth_count)

plt.ylabel('No. of live births')
plt.title('Live births over the years')
plt.show()
```

```python
data = np.genfromtxt("singapore-residents-by-ethnic-group-and-sex-end-june-annual.csv",
            delimiter=',',
            names=True, dtype=(int, 'U22', int))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 10))
data_1960  = data[data['year'] == 1960]
data_1960_male = data_1960[data_1960['level_1'] == "Total Male Residents"]['value'][0]
data_1960_female = data_1960[data_1960['level_1'] == "Total Female Residents"]['value'][0]

labels = ["Total Male residents", "Total Female residents"]
axes[0].pie([data_1960_male, data_1960_female], labels=labels, autopct='%1.1f%%')
axes[0].set_title("Gender composition in 1960")


# Divider
data_2016  = data[data['year'] == 2016]
data_2016_male = data_2016[data_1960['level_1'] == "Total Male Residents"]['value'][0]
data_2016_female = data_2016[data_1960['level_1'] == "Total Female Residents"]['value'][0]

labels = ["Total Male residents", "Total Female residents"]
axes[1].pie([data_2016_male, data_2016_female], labels=labels, autopct='%1.1f%%')
axes[1].set_title("Gender composition in 2016")
fig.tight_layout()
```

```python
# •	surface-air-temperature-monthly-mean
data = np.genfromtxt("surface-air-temperature-monthly-mean.csv",
            delimiter=',',
            names=True, 
            dtype=None)
mean_temperature = data["mean_temp"]

data = np.genfromtxt("rainfall-monthly-number-of-rain-days.csv",
            delimiter=',',
            names=True, 
            dtype=None)

no_of_rainy_days = data['no_of_rainy_days']
plt.scatter(mean_temperature, no_of_rainy_days, color="green")
plt.title('Monthly rainfall and temperatures since Jan 1982')
plt.xlabel('Mean monthly temperatures')
plt.ylabel('Number of rainy days')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(mean_temperature, no_of_rainy_days, 1)

#add linear regression line to scatterplot 
plt.plot(mean_temperature, m*mean_temperature+b, color="red")
```

```python
data = np.genfromtxt("coe-results.csv",
            delimiter=',',
            names=True, dtype=('U7',int,'U10',int,int,int,int))

#set up the y values
catA = data[data['vehicle_class'] == 'Category A']['quota']
catB = data[data['vehicle_class'] == 'Category B']['quota']
catC = data[data['vehicle_class'] == 'Category C']['quota']
catD = data[data['vehicle_class'] == 'Category D']['quota']
catE = data[data['vehicle_class'] == 'Category E']['quota']
y_values = np.array([catA, catB, catC, catD, catE])

#set up the labels on the x-axis
x_labels = np.unique(data['vehicle_class'])
x_ticks = np.arange(len(x_labels))
plt.xticks(rotation=90)

plt.boxplot(y_values.transpose(),labels=x_labels, patch_artist=True)
plt.ylabel("Quota")
plt.title("Quota of COE for different vehicle classes")
plt.show()
```

```python
# •	median-rentals-and-vacancy-of-retail-space-by-locality-quarterly.csv

data = np.genfromtxt("median_rentals-and-vacancy.csv",
            delimiter=',',
            names=True, dtype=("U7", "U30", float, float, float))

plt.figure(figsize=(8, 8))

#set up the y values
data_orchard = data[data["locality"] == "Orchard"]["ret_med_rent_lease_cm"]
data_outside_central_area = data[data["locality"] == "Outside Central Area"]["ret_med_rent_lease_cm"]
data_central_area_outside_orchard = data[data["locality"] == "Central Area - Outside Orchard"]["ret_med_rent_lease_cm"]
y_values = np.array([data_orchard, data_outside_central_area, data_central_area_outside_orchard])

#set up the labels on the x-axis
x_labels = ["Orchard", "Outside Central Area", "Central Area"]
x_ticks = np.arange(len(x_labels))

plt.boxplot(y_values.transpose(), labels = x_labels, sym='g')
plt.ylabel("$psf")
plt.title("Median Rentals in Orchard, Outside Central Area and Central Area")
plt.show()
```
