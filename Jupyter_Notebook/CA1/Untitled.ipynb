```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("coe-results.csv",
                    delimiter=",",
                    names=True, dtype=('U7', int, 'U10', int, int, int, int))
```

# Section 4

```python
print(f"***** COE Results *****")
print(f"There are {len(data)} in this dataset")

data_months = data['month']
months = np.unique(data_months)
print(f"There are {len(months)} months of data captured from {months[0]} to {months[-1]}")
print(f"There are {len(np.unique(data['vehicle_class']))} vehicle classes")

# Getting the lowest and highest premium of CAT A
keyword = ["Category A", "Category B", "Category C", "Category D", "Category E"]
for i in range(len(keyword)):
    # keywords
    keyword_specific = keyword[i];
    # specfics dataframes relative to keyword
    data_category_specific = data[data['vehicle_class'] == keyword_specific];
    # maxes
    data_category_specific_max = data_category_specific['premium'].max();
    data_category_specific_argmax = data_category_specific['premium'].argmax();
    # mins
    data_category_specific_min = data_category_specific['premium'].min();
    data_category_specific_argmin = data_category_specific['premium'].argmin();
    # printing
    print(f'\nThe lowest premium of {keyword_specific} COE is {data_category_specific_min}, on month {data_category_specific[data_category_specific_argmin]["month"]}, bid {data_category_specific[data_category_specific_argmin]["bidding_no"]}.')
    print(f'The highest premium of {keyword_specific} COE is {data_category_specific_max}, on month {data_category_specific[data_category_specific_argmax]["month"]}, bid {data_category_specific[data_category_specific_argmax]["bidding_no"]}.')
```

```python
data = np.genfromtxt("health-facilities-primary-care-dental-clinics-and-pharmacies.csv",
                    delimiter=",",
                    names=True, dtype=(int, 'U30', 'U10', 'U20', int))

plt.figure(figsize=(20, 6))

plt.subplot(121)
plt.hist(data[data['sector'] == "Public"]['no_of_facilities'])
plt.title('Histogram of the dental public sector since 2009')
plt.xlabel('')
plt.ylabel('No. of public health clinics')

plt.subplot(122)
plt.hist(data[data['sector'] == "Private"]['no_of_facilities'])
plt.title('Histogram of the dental private sector since 2009')
plt.xlabel('')
plt.ylabel('No. of private health clinics')
```

```python
#number-of-dentists.csv
data = np.genfromtxt("number-of-dentists.csv",
                    delimiter=",",
                    names=True, dtype=(int, 'U30', 'U30',int))

plt.figure(figsize=(20, 6))

plt.subplot(121)
plt.hist(data[data['sector'] == "Public Sector"]['count'])
plt.title('Histogram of the dental public sectors since 2009')
plt.xlabel('')
plt.ylabel('No. of public Specialists')

plt.subplot(122)
plt.hist(data[data['sector'] == "Private Sector"]['count'])
plt.title('Histogram of the dental private sectors since 2009')
plt.xlabel('')
plt.ylabel('No. of private Specialists')
```

```python
#number-of-dentists.csv
data = np.genfromtxt("dental-index-dental-health-status-of-the-school-children-at-12-and-15-years-old.csv",
                    delimiter=",",
                    names=True, dtype=(int, 'U30', float))

plt.figure(figsize=(20, 6))

plt.subplot(131)
plt.hist(data[data['age'] == "12 YEARS OLD"]['dental_index'])
plt.title('Histogram of the health index 12 YOs since 2003')
plt.xlabel('')
plt.ylabel('No. of dental_index')

plt.subplot(132)
plt.hist(data[data['age'] == "15 YEARS OLD"]['dental_index'])
plt.title('Histogram of the health index of 15 YOs since 2003')
plt.xlabel('')
plt.ylabel('No. of dental_index')

years = np.unique(data['year'])
data_pool = []
plt.subplot(133)
for i in range(len(years)):
    data_pool.append(data[data['year'] == years[i]]['dental_index'].sum())
    
plt.hist(data_pool)
plt.title('Histogram of the health index of 15 YOs since 2003')
plt.xlabel('years')
plt.ylabel('No. of dental_index')
```
