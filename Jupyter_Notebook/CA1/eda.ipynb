```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
dental = pd.read_csv('dental-index-dental-health-status-of-the-school-children-at-12-and-15-years-old.csv')
dental.head()
```

```python
dental
```

```python
yo12 = dental[dental['age'] == '12 YEARS OLD']
yo15 = dental[dental['age'] == '15 YEARS OLD']
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
# Plotting out for 12 years old
line1 = ax.plot(yo12['year'], yo12['dental_index'])

# Plotting out for 15 years old
line2 = ax.plot(yo15['year'], yo15['dental_index'])

plt.show()
```
